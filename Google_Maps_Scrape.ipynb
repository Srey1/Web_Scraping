## Extract the name, rating, phone number, and url of each buisness

```python
names = soup.find_all("div", class_ = "rgnuSb")
text_names = [name.text for name in names]
ratings = soup.find_all("div", class_ = "rGaJuf")
text_ratings = [rating.text for rating in ratings]
phone_numbers_container = soup.find_all("div", class_ = "I9iumb")
phone_numbers_container = phone_numbers_container[2::3]
#print(phone_numbers_container[4])
phone_numbers = [container.find_all("span", class_ = "hGz87c")[-1].text.strip() for container in phone_numbers_container]
large_url_container = soup.find_all("div", class_ = "SJyhnc")
first_url = (large_url_container[2])
first_url.find_all("div", {"data-website-url": True})
final_url = []
for container in large_url_container:
    url = container.find_all("div", {"data-website-url": True})
    if url == []:
        final_url.append("Unknown")
    else:
        final_url.append(url[0]["data-website-url"])
text_names
phone_numbers
#urls = [url["data-website-url"] for url in url_container]
```

## Create the pandas dataframe

```python
import pandas as pd
df = pd.DataFrame(columns = ["Name", "Rating", "Phone Number", "URL"])
df
```
